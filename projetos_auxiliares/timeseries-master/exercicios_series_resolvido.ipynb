# Exercício de séries temporais
1. Analíse as tendências das pesquisas sobre academia e finanças
2. Analíse as sazonalidades das pesquisas sobre academia e finanças

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib qt
```

```python
dados = pd.read_csv('timeline.csv')
```

```python
dados.month = pd.to_datetime(dados.month)
```

```python
dados.set_index('month', inplace=True)
```

```python
gym = dados.gym
finance = dados.finance
```

```python
gym.rolling(12).mean().plot()
```

Existe uma tendência linear de crescimento na pesquisa sobre academia.

```python
finance.rolling(12).mean().plot()
```

```python
gym.diff(1).plot()
```

```python
intervalo = (dados.index >= '2005-01') & (dados.index < '2008-01') 
gym.loc[intervalo].diff().plot()
```

Sobre a sazonalidade das pesquisa relacionadas a academia, observamos um pico sazonal em janeiro, seguido de uma queda em fevereiro com novo pico em agosto, finalizando uma mais uma queda em outubro ou novembro.

```python
intervalo = (dados.index >= '2006-01') & (dados.index < '2017-01') 
finance.loc[intervalo].diff().plot()
```

Para as pesquisas sobre finanças, observamos a ausência de comportamento sazonal.
