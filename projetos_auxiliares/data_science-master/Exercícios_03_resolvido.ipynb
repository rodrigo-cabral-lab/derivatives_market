# Exercícios: Parte 03
## Verifique as seguintes indagações através da construção de histogramas ou gráficos de frequência:
1. Qual tipo de imóvel possui maior frequência?
2. Qual quantidade de banheiros que aparece com maior frequência para cada tipo de imóvel?
3. A maior parte dos apartamentos estão longe ou perto do aeroporto?
4. Existem mais apartamentos velhos ou novos? 
5. As casas estão concentradas mais ao norte ou ao sul?

```python
import pandas as pd
import matplotlib.pyplot as plt

housing = pd.read_csv('https://raw.githubusercontent.com/ronaldolagepessoa/data_science/master/dados/housing_ok.csv')
```

### Solução - 1
As casas são o tipo de imóvei com maior frequência.

```python
tipo_freq = housing.tipo.value_counts()
plt.bar(tipo_freq.index, tipo_freq.values)
```

### Solução - 2
As casas e kitnets possuem, em sua maoria, 1 banheiro. Já os apartamentos, em sua maioria, possuem 2 banheiros.

```python
filtro = housing.tipo == 'kitnet'
banheiros_freq = housing[filtro].banheiros.value_counts()
plt.title('Gráfico de frequência')
plt.xlabel('Número de quartos')
plt.ylabel('Frequência')
plt.bar(banheiros_freq.index, banheiros_freq.values)
```

```python
filtro = housing.tipo == 'casa'
banheiros_freq = housing[filtro].banheiros.value_counts()
plt.title('Gráfico de frequência')
plt.xlabel('Número de quartos')
plt.ylabel('Frequência')
plt.bar(banheiros_freq.index, banheiros_freq.values)
```

```python
filtro = housing.tipo == 'apartamento'
banheiros_freq = housing[filtro].banheiros.value_counts()
plt.title('Gráfico de frequência')
plt.xlabel('Número de quartos')
plt.ylabel('Frequência')
plt.bar(banheiros_freq.index, banheiros_freq.values)
```

### Solução - 3
Quanto maior a distância do aeroporto, maior e a concentração de apartamentos.

```python
filtro = housing.tipo == 'apartamento'
plt.hist(housing[filtro].distancia_aeroporto)
```

### Solução - 4
Existem mais apartamentos novos.

```python
filtro = housing.tipo == 'apartamento'
plt.hist(housing[filtro].ano_construcao)
```

### Solução - 5
Existe baixa concentração de casas no extremo sul da cidade.

```python
filtro = housing.tipo == 'casa'
plt.hist(housing[filtro].latitude)
```
