## Aplicando modelos na bolsa de valores



```python
import numpy as np
import pandas_datareader.data as web

final = np.datetime64('today')
inicio = final - np.timedelta64(90, 'D')

df = web.DataReader(['PETR4.SA', 'VALE3.SA', 'VVAR3.SA'], 'yahoo', inicio, final).Close
df.head()
```

```python
df.plot(figsize=(15, 5))
```

```python
from statsmodels.tsa.arima_model import ARIMA 

model1 = ARIMA(df['PETR4.SA'], order=(5, 1, 2)).fit()
```

```python
model2 = ARIMA(df['VALE3.SA'], order=(5, 1, 2)).fit()
```

```python
model3 = ARIMA(df['VVAR3.SA'], order=(5, 1, 2)).fit()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))
plt.plot(df['PETR4.SA'])
plt.plot(model1.predict(typ='levels'))
plt.legend(['real', 'previsto'])
```

```python
print(f'PETR4 = {model1.forecast()[0][0]}')
print(f'VALE3 = {model2.forecast()[0][0]}')
print(f'VVAR3 = {model3.forecast()[0][0]}')
```

```python
print('Incremento PETR4 = {}'.format(model1.forecast()[0][0] - df['PETR4.SA'].values[-1]))
print('Incremento VALE3 = {}'.format(model2.forecast()[0][0] - df['VALE3.SA'].values[-1]))
print('Incremento VVAR3 = {}'.format(model3.forecast()[0][0] - df['VVAR3.SA'].values[-1]))
```
