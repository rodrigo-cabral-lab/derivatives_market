# Exercícios: Parte 01
## Verifique as seguintes informações:
1. Compare o crescimento populacional dos Estados Unidos, Brazil e Argentina;
2. Compare o crescimento populacional entre os países da América Latina
3. Compare o crescimento relativo entre os países da América Latina

## Solução:

### Questão 1:
Primeiro faremos a importação dos pacotes ```pandas``` e ```matplotlib.pyplot```

````python
import pandas as pd
import matplotlib.pyplot as plt
````

Em seguida, faremos a importação da tabela ```countries.csv``` utilizando o comando ```pd.read_csv()```.

````python
countries = pd.read_csv('https://raw.githubusercontent.com/ronaldolagepessoa/data_science/master/dados/countries.csv')
````

Agora precisamos filtrar as informações dos três países

````python
usa = countries[countries.country == 'United States'].copy()
br = countries[countries.country == 'Brazil'].copy()
ar = countries[countries.country == 'Argentina'].copy()
````

Por último faremos a plotagem dos crescimentos populacionais

````python
plt.plot(usa.year, usa.population)
plt.plot(br.year, br.population)
plt.plot(ar.year, ar.population)
plt.legend(['USA', 'BR', 'AR'])
````

### Questão 2
Para comparar o crescimento populacional dos países da américa do sul precisamos, primeiramente, listar todos eles:
* Brazil
* Argentina
* Colombia
* Paraguay
* Uruguay
* Chile
* Bolivia
* Venezuela

Em seguida faremo os filtros para cada país.

````python
br = countries[countries.country == 'Brazil'].copy()
ar = countries[countries.country == 'Argentina'].copy()
co = countries[countries.country == 'Colombia'].copy()
pa = countries[countries.country == 'Paraguay'].copy()
ur = countries[countries.country == 'Uruguay'].copy()
ch = countries[countries.country == 'Chile'].copy()
bo = countries[countries.country == 'Bolivia'].copy()
ve = countries[countries.country == 'Venezuela'].copy()
````

Por fim, plotamos as informações desejadas

````python
plt.plot(br.year, br.population)
plt.plot(ar.year, ar.population)
plt.plot(co.year, co.population)
plt.plot(pa.year, pa.population)
plt.plot(ur.year, ur.population)
plt.plot(ch.year, ch.population)
plt.plot(bo.year, bo.population)
plt.plot(ve.year, ve.population)
plt.legend(['Brasil', 'Argentina', 'Colembia', 'Paraguay', 'Uruguay', 'Chile', 'Bolivia', 'Venezuela'])
````

### Questão 3
Para comparar os crescimentos percentuais precisamos criar uma nova coluna. A nova coluna conterá os valores da coluna "population" dividida pela primeira população (multiplicado por 100 para ficar em percentual).

````python
br['population_perc'] = br.population / br.iloc[0, 2] * 100
ar['population_perc'] = ar.population / ar.iloc[0, 2] * 100
co['population_perc'] = co.population / co.iloc[0, 2] * 100
pa['population_perc'] = pa.population / pa.iloc[0, 2] * 100
ur['population_perc'] = ur.population / ur.iloc[0, 2] * 100
ch['population_perc'] = ch.population / ch.iloc[0, 2] * 100
bo['population_perc'] = bo.population / bo.iloc[0, 2] * 100
ve['population_perc'] = ve.population / ve.iloc[0, 2] * 100
````

Por fim, plotaremos as informações

````python
plt.plot(br.year, br.population_perc)
plt.plot(ar.year, ar.population_perc)
plt.plot(co.year, co.population_perc)
plt.plot(pa.year, pa.population_perc)
plt.plot(ur.year, ur.population_perc)
plt.plot(ch.year, ch.population_perc)
plt.plot(bo.year, bo.population_perc)
plt.plot(ve.year, ve.population_perc)
plt.legend(['Brasil', 'Argentina', 'Colembia', 'Paraguay', 'Uruguay', 'Chile', 'Bolivia', 'Venezuela'])
````
