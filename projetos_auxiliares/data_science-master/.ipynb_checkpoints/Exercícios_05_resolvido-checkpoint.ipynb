# Exercícios: Parte 05
1. Verifique a distribuição de preços para cada cada quantidade de quartos

    1.1. de todos os imóveis;
    
    1.2. das casas;
    
    1.3. dos kitnets;
    
    1.4. dos apartamentos.
2. Verifique a distribuição dos anos de construção para os três tipos de imóvel.
3. Para os apartamentos, verificar a distribuição do ano de construção para cada quantidade de quartos.

```python
import pandas as pd
import matplotlib.pyplot as plt

housing = pd.read_csv('https://raw.githubusercontent.com/ronaldolagepessoa/data_science/master/dados/housing_ok.csv')
```

### Solução - 1
Tanto para o caso geral como para as casas, o preço cresce tanto na média como na variabilidade até 5 quartos. Com seis quartos, o preço médio se mantem aproximadamente o mesmo mas com uma redução na variabilidade. Com sete quartos, o preço médio cai e volta ao valor máximo com oito quartos.  

Para os kitnets, os preços aumentam em média e variabilidade com o aumento do número de quartos. O mesmo é observado para os apartamentos, com a diferença na diminuição do preço para cinco cinco quartos.

```python
housing.boxplot(column='preco', by='quartos', showfliers=False)
```

```python
housing[housing.tipo == 'casa'].boxplot(column='preco', by='quartos', showfliers=False)
```

```python
housing[housing.tipo == 'kitnet'].boxplot(column='preco', by='quartos', showfliers=False)
```

```python
housing[housing.tipo == 'apartamento'].boxplot(column='preco', by='quartos', showfliers=False)
```

### Solução - 2
O ano de construção dos apartamentos estão entre os mais recentes, onde boa parte das unidades possuem ano de construção maior o igual a 2000. As casas possuem grande variabilidade no ano de construção, com valor mediano igual 1950. Os kitnets possuem variabilidade maior que os apartamentos, com mediana pouco acima de 1975. 

```python
housing.boxplot(column='ano_construcao', by='tipo', showfliers=False)
```

### Solução - 3
Segundo o gráfico, a maioria dos apartamentos com um quarto possuem ano de construção entre 1940 e 1960. Para dois ou mais quartos, a mediana gira em torno do ano de 2005, onde a variabilidade dos valores diminui com o aumento do número de quartos.

```python
housing[housing.tipo == 'apartamento'].boxplot(column='ano_construcao', by='quartos', showfliers=False)
```
