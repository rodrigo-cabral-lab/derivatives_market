# Case 1 - Aluguel/Vendas de imóveis em São Paulo/SP

Para praticar os comandos apreesentados nos últimos notebooks, exploraremos as informações contidas no arquivo ```sao-paulo-properties-april-2019.csv``` com o objetivo de responder às seguintes indagações:

1. Como se distribui os preços dos alugueis?
2. Como se distribui os preços dos imóveis?
3. Para os alugueis, quais as 10 maiores frequência de ofertas, considerando o número de quartos, suites e banheiros?
4. Qual o preço médio dos imóveis novos e usados, mobiliados e não mobiliados?
5. Qual a distribuição de preços considerando os mesmos atributos anteriores (4)?
6. Verifique se existe relação entre o preço dos imóveis e o valor do comdomínio.
7. Verifique se existe relação entre o valor do aluguel e o valor do comdomínio.
8. Plote as coordenadas geográficas identificando a região que possui os maiores preços de aluguel.
9. Identifique os 10 distritos com os maiores preços médios de aluguel e os 10 distritos com os menores preços médios de aluguel.

````python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('dados/sao-paulo-properties-april-2019.csv')
df.head()
````

````python
df.info()
````

Temos um total de 13640 linhas e 16 colunas sendo, sendo duas do tipo ```float64```, onze do tipo ```int64``` e três do tipo ```object```. 

Primeiro, faremos o tratamento das informações iniciando pela renomeação das colunas. Acessando o parâmetro ```DataFrame.columns```, temos acesso a lista de colunas presentes no ```DataFrame```.

````python
df.columns
````

Faremoss a cópia dos nomes para criar o dicionário que será utilizado no comando ```DataFrame.rename()``` para alteração dos nomes das colunas. 

Os novos nomes serão:
* 'Price': 'preco'
* 'Condo': 'condominio'
* 'Size': 'tamanho'
* 'Rooms': 'quartos' 
* 'Toilets': 'banheiros'
* 'Suites': 'suites'
* 'Parking': 'vagas'
* 'Elevator': 'elevador'
* 'Furnished': 'mobiliado'
* 'Swimming Pool': 'piscina' 
* 'New': 'novo'
* 'District': 'distrito'
* 'Negotiation Type': 'negociacao'
* 'Property Type': 'tipo'
* 'Latitude': 'latitude' 
* 'Longitude': 'longitude'

````python
df.rename(columns={'Price': 'preco', 'Condo': 'condominio', 'Size': 'tamanho', 'Rooms': 'quartos', 
           'Toilets': 'banheiros', 'Suites': 'suites', 'Parking': 'vagas',
       'Elevator': 'elevador', 'Furnished': 'mobiliado', 'Swimming Pool': 'piscina', 
           'New': 'novo', 'District': 'distrito',
       'Negotiation Type': 'negociacao', 'Property Type': 'tipo', 'Latitude': 'latitude', 
           'Longitude': 'longitude'}, inplace=True)
````

````python
df.columns
````

As colunas ```elevador```, ```mobiliado```, ```piscina``` e ```novo``` são colunas binárias, como visto abaixo com o comando ```Series.unique()```.

````python
print(df.elevador.unique())
print(df.mobiliado.unique())
print(df.piscina.unique())
print(df.novo.unique())
````

Modificaremos os valores binário por "sim" e "nao" utilizando o comando ```Series.replace()```.

````python
df.elevador.replace({0: 'nao', 1: 'sim'}, inplace=True)
df.mobiliado.replace({0: 'nao', 1: 'sim'}, inplace=True)
df.piscina.replace({0: 'nao', 1: 'sim'}, inplace=True)
df.novo.replace({0: 'nao', 1: 'sim'}, inplace=True)
````

````python
df.head()
````

Por fim, utilizaremos o comando ```DataFrame.describe()``` para insvestigar possíveis incoerências nos dados.

````python
df.describe()
````

Observamos uma incoerência associada às coordenadas geográficas. O valores máximos da latitude e longitude são iguais a zero e o valores mínimos estão distantes das medianas, sugerindo erros de cadastro.

Utilizaremos como referência os valores da mediana para filtrar as coordenadas corretas. Para tanto, faremos uso do comando ```Series.between()```.

````python
df = df[df.longitude.between(-48, -44)]
df = df[df.latitude.between(-25, -22)]
````

## Resposta (1)
Para responder à perguna (1), criaremos um ```DataFrame``` contendo somente os alugueis e outro contendo somente os preços de venda.

````python
df_rent = df[df.negociacao == 'rent']
df_sale = df[df.negociacao == 'sale']
````

Utilizaremos, primeiro, o histograma para analisar a distribuição de valores de aluguel. Para tanto, utilizaremos o comando ```Series.hist()```.

````python
df_rent.preco.hist(figsize=(15, 5))
````

Notamos a presença de _outliers_ nos valores de alugueis. Para visualizar a distribuição de valores sem a presença de _outliers_ utilizaremos um filtro para eliminar valores acima de 6000.

````python
df_rent[df_rent.preco <= 6000].preco.hist(figsize=(15, 5))
````

Observamos, agora, que os valores de aluguel se concentram no intervalo entre 1000 a 3200, aproximadamente. È possível identificar esse padrão através do gráfico boxplot. Para tanto, utilizaremos o comando ```Series.plot(kind='box', showfliers=False)```.

````python
df_rent.preco.plot(kind='box', showfliers=False)
````

## Resposta (2)

Utilizaremos o mesmo procedimento para analisar os preços dos imóveis.

````python
df_sale[df_sale <= 2e6].preco.hist(figsize=(15, 5))
````

````python
df_sale.preco.plot(kind='box', showfliers=False)
````

Tanto o histograma como o boxplot mostram alta frequência no intervalo de preços entre 250 mil e 750 mil reais.

## Resposta (3)

Utilizaremos o comando ```DataFrame.groupby()```, passando como parâmetro de agrupamento uma lista contendo os nomes das colunas ```quartos```, ```suites``` e ```banheiros```. Como faremos a contagem da frequência, podemos escolher qualquer coluna de referência. Nesse caso utilizaremos a coluna ```tipo```, seguido do comando ```count```. Ordenaremos a ```Series``` resultante em ordem decrescente com o comando ```sort_values(ascending=False)``` e faremos a seleção das 10 primeiras colunas com o comando ```head(10)```. Para analisar o resultado de maneira visual, finalizaremos com o comando ```plot(kind='bar'```, gerando um gráfico de barras com as frequências de cada combinação de atributos.

````python
df_rent.groupby(['quartos', 'suites', 'banheiros']).tipo.count().sort_values(ascending=False).head(10).plot(kind='bar')
````

## Resposta (4)

Para comparar a média de preço para os atributos ```novo``` e ```mobiliado```, utilizamos o comando ```DataFrame.groupby()``` aplicando a média na coluna ```preco```. Acrescentamos o comando ```plot(kind='bar')``` para mostrar o restado no formato de um gráfico de barras.

````python
df_sale.groupby(['novo', 'mobiliado']).preco.mean().plot(kind='bar')
````

### Resposta (5)

Para analisar a distribuição dos preços para cada combinção de valores dos atributos ```novo``` e ```mobiliado```, utilizamos o comando ```DataFrame.boxplot()```. Para o parâmetro ```column``` passaremos a coluna ```preco``` e, para o parâmetro ```by``` passaremos a lista contendo as colunas ```novo``` e ```mobiliado```. Acrescendaremos, também, o parâmetro ```showfliers=False``` para eliminação automática dos _outliers_.

````python
df_sale.boxplot(column='preco', by=['novo', 'mobiliado'], showfliers=False)
````

Os preços dos imóveis usados mobiliados possuam a maior variabilidade, seguido dos imóveis usados não mobiliados e dos imóveis novos não mobiliados. Os imóveis novos mobilidados possuam baixa variabilidade nos preços, provavelmente pela baixa ocorrência dessa configuração.

Para confirmar a baixa frequência de imóveis novos mobiliados, filtraremos as instâncias com essas características.

````python
df_sale[(df_sale.novo == 'sim') & (df_sale.mobiliado == 'sim')]
````

Como esperado, temos somente dois imóveis com essa configuração.

Outra forma de analisar a variabilidade dos preços dos imóveis pe através do agrupamento de dados com o cálculo do desvio padrão. Para tanto, passamos como método de agrupameto o comando ```std()```.

````python
df_sale.groupby(['novo', 'mobiliado']).preco.std().plot(kind='bar')
````

## Resposta (6)

Para analisar a relação entre o valor do condomínio e o preço do imóvel utilizamos o gráfico de dispersão (_scatter plot_) através do comando ```DataFrame.plot(kind='scatter')```. Para o parâmetro ```x```, passaremos a coluna ```condominio``` e, para o parâmetro ```y```, passaremos a coluna ```preco```.

````python
df_sale.plot(kind='scatter', x='condominio', y='preco')
````

Parece haver um problema em relação aos condomínios com valor igual a zero. Eliminaremos esses valores utilizando um filtro.

````python
df_sale[df_sale.condominio > 0].plot(kind='scatter', x='condominio', y='preco')
````

Percebemos um padrão linear entre o valor do condomínio e o preço do imóvel.

## Resposta (7)

Utilizaremos os mesmos comando anteriores aplicados ao ```DataFrame``` ```df_rent```.

````python
df_rent[df_rent.condominio > 0].plot(kind='scatter', x='condominio', y='preco')
````

### Resposta (8)

Para criar o gráfico das coordenadas utilizaremos o comando ```DataFrame.plot()```. Para os parâmetros ```x``` e ```y```, passaremos as colunas ```longitude``` e ```latitude```, respectivamente. Para o parâmetro ```c```, passaremos a coluna ```preco```. Para manter as escalas dos dois eixos iguals, importaremos o pacote ```pyplot``` e utilizaremos o comando ```pyplot.axis('equal')```.

````python
import matplotlib.pyplot as plt

filtro = (df_rent.condominio > 0) & (df_rent.preco < 6000)
df_rent[filtro].plot(kind='scatter', x='longitude', y='latitude', c='preco', cmap='magma', figsize=(17, 10))
plt.axis('equal')
````

````python
df_rent.groupby('distrito').preco.mean().sort_values(ascending=False).head(10).plot(kind='bar', rot=45, figsize=(15, 6), fontsize=12)
````
