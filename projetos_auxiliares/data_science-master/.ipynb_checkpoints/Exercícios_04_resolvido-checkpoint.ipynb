# Exercícios: Parte 04
## Responda às indagações abaixo utilizando o gráfico _scatter_:
1. Existe relação entre o preço a área do terreno dos apartamentos?
2. Plote as coordenadas das casas, kitnets e apartamentos para verificar se existem regiões com predominância de cada tipo.
3. Utilize o gráfico _scatter_ para verificar a posição aproximada do aeroporto. Existem imóveis com a informação incorreta da distância do aeroporto?
4. Descubra qual a parte mais antiga da cidade e como se deu o crescimento da mesma.

```python
import pandas as pd
import matplotlib.pyplot as plt

housing = pd.read_csv('https://raw.githubusercontent.com/ronaldolagepessoa/data_science/master/dados/housing_ok.csv')
```

### Solução - 1
Por conta da presença de _outliers_, consideramos somente as áreas de terrenos menores ou iguais a mil metros quadrados.

__Conclusão__: Não parece existir relação entre os dois atributos.

```python
filtro = (housing.tipo == 'apartamento') & (housing.area_terreno <= 1000)
plt.scatter(housing[filtro].area_terreno, housing[filtro].preco)
```

```python
filtro = (housing.tipo == 'casa') & (housing.preco <= 5000000)
plt.scatter(housing[filtro].latitude, housing[filtro].preco)
```

### Solução - 2
Existem pequenas regiões com concentração de casas.

```python
casas = housing[housing.tipo == 'casa']
kitnets = housing[housing.tipo == 'kitnet']
apartamentos = housing[housing.tipo == 'apartamento']

plt.figure(figsize=(15, 10))
plt.axis('equal')
plt.scatter(casas.longitude, casas.latitude)
plt.scatter(kitnets.longitude, kitnets.latitude)
plt.scatter(apartamentos.longitude, apartamentos.latitude)
plt.legend(['casa', 'kitnet', 'apartamento'])
```

### Solução - 3
O aeroporto parece estar localizado na área central da cidade. Alguns pontos parecem estar com a informação da distância do aeroporto errada.

```python
plt.figure(figsize=(15, 10))
plt.axis('equal')
plt.scatter(housing.longitude, housing.latitude, c=housing.distancia_aeroporto)
plt.colorbar()
```

### Solução - 4
A área central da cidade é mais antiga que a periferia.

```python
plt.figure(figsize=(15, 10))
plt.axis('equal')
plt.scatter(housing.longitude, housing.latitude, c=housing.ano_construcao)
```
