## Parte 04: Gráfico de dispersão (scatter)

Gráficos scatter são utilizados para comparar pares de valores com o objetivo de identificar relações de dependência. Na presença de coordenadas geográficas, o gráfico _scatter_ pode ser utilizado para visualização da distribuição de pontos.

Utilizaremos a base ```housing_ok.csv``` como exemplo.

````python
import pandas as pd
import matplotlib.pyplot as plt

housing = pd.read_csv('https://raw.githubusercontent.com/ronaldolagepessoa/data_science/master/dados/housing_ok.csv')
housing.quartos = housing.quartos.astype('int8')
housing.banheiros = housing.banheiros.astype('int8')
housing.vagas = housing.vagas.astype('int8')
````

## Plotando coordenadas geográficas
Podemos, inicialmente, plotar o gráfico scatter para termos uma ideia da distribuição geográfica dos imóveis. Para isso utilizaremos a função [```DataFrame.plot.scatter()```](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.scatter.html). Os parâmetros ```x``` e ```y``` recebem o nome da coluna contendo os dados para cada um dos respectivos eixos. Definiremos, também, o tamanho da figura para as dimensões $10\times 8$ através do parâmetro ```figsize=(10, 0)```. Para manter idêntica a escala dos eixos, acrescentamos o comando [```plt.axis('equal')```](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.axis.html).

````python
housing.plot.scatter(x='longitude', y='latitude', figsize=(10, 8))
plt.axis('equal')
````

### Escala de cores
Para melhorar a quantidade de informação mostrada no gráfico scatter, podemos adicionar escala de cores e de forma.

No exemplo, utilizaremos a escala de cores para representar o preço dos imóveis. Definiremos o parâmetro ```c='preco'```. Por padrão, a escala de cor está é definida como a do cinza. Para mudar a escala de cor, alteramos o valor do parâmetro ```cmap```. Abaixo, temos algumas das opções disponíveis:
* ```viridis```
* ```plasma```
* ```inferno```
* ```magma```
* ```cividis```

````python
housing.plot.scatter(x='longitude', y='latitude', c='preco', cmap='magma', figsize=(10, 8))
plt.axis('equal')
````

### Escala de tamanho
Para utilzar a escala de tamanho definimos o parâmetro ```s=DataFrame.nome_da_coluna```. 
Para o exemplo, utilizaremos o parâmetro de forma para visualizar o tamanho dos imóveis através da área construida. Dividiremos a área contruida por $5$ para reduzir o tamanho dos pontos.

````python
housing.plot.scatter(x='longitude', y='latitude', c='preco', cmap='magma', s=housing.area_construida / 5, figsize=(10, 8))
plt.axis('equal')
````

### Transparência dos pontos
Ainda assim, por conta da quantidade de pontos, a visualização das informações não está perfeita. Podemos, ainda, definir o parâmetro ```alpha``` que define a transparência dos pontos mostrados no gráfico. O valor padrão do parâmetro é igual a 1.0 (100%). Para o exemplo, faremos ```alpha=0.6```.

````python
housing.plot.scatter(x='longitude', y='latitude', c='preco', cmap='magma', s=housing.area_construida / 5, figsize=(10, 8), alpha=0.6)
plt.axis('equal')
````

## Verificando relação entre atributos da tabela
O gráfico _scatter_ pode ser utilizado para verificar a existência de relação entre colunas (variáveis) contidas em um ```DataFrame```.  No exemplo, verificaremos a relação entre a área construída e o preço do imóvel.

````python
housing.plot.scatter(x='area_construida', y='preco')
````

### Filtrando informações para facilitar a análise
É possível perceber que existe uma relação forte entre a área construída e o preco dos imóveis (possivelmente linear).

Existem, porém, alguns pontos (poucos) que não obedecem o padrão geral observado. Chamamos esses pontos de _outliers_ (pontos fora da curva).
Os _outliers_ podem atrapalhar a análise de correlação e devem ser eliminados. Para tanto, acrescentaremos um filtro para eliminar valores de área contruída maiores que 500 e preços maiores que 5 milhões. Para criar filtros com mais de uma condição, utilizamos a notação ```(condição_1) & (condição_2)```. O símbolo ```&``` é utilizado para explicitar que as duas condições devem ser respeitadas simultaneamente.

````python
filtro = (housing.area_construida <= 500) & (housing.preco <= 5000000)
housing[filtro].plot.scatter(x='area_construida', y='preco')
````

Ainda assim, observamos um padrão linear de relação entre o preço dos imóveis e a área construída dos mesmos.

## Gráfico Hexabin
Dentre os gráfico disponíveis em um ```DataFrame``` temos o [```DataFrame.plot.hexbin()```](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.hexbin.html). O gráfico hexabin possui a mesma lógica do gráfico scatter, com a diferença de acrescentar uma medida de consentração de pontos através de uma escala de cores. Utilizar esse gráfico é importante para melhor entender a distribuição dos dados.

No exemplo abaixo, simplesmente substituimos a função ```DataFrame.plot.scatter()``` por ```DataFrame.plot.hexbin()```.

````python
filtro = (housing.area_construida <= 500) & (housing.preco <= 5000000)
housing[filtro].plot.hexbin(x='area_construida', y='preco', cmap='magma', figsize=(10, 6))
````

Com conclusão, observamos que a relação linear entre preço e imóvel é mais confiável para imóveis com preço abaixo de 2 milhão e área contruída menor que 200 metros quadrados. Dessa forma, seria prudente filtrar os valores fora desses intervalos em uma eventual contrução de um modelo preditivo de regressão linear.
