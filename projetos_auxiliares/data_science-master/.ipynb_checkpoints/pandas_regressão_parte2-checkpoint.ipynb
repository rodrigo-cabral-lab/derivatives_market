# Regressão linear múltipla

Regressão linear múltipla possui a mesma lógica da regressão linear simples, diferenciando-se somente no número de variáveis independentes.

Uma regressão linear múltipla possui múltiplas variáveis independentes $x$, i.e., $x_1, x_2,$... Assim, podemos representar a equação que prevê o valor de $y$ em função de cada $x_i$ como:
$$y=\alpha_1 x_1 + \alpha_2 x_2 +\cdots \alpha_n x_n + \beta$$

Para exemplificar, considere os dados $y$ seguindo um comportamento linear dependente de $x1$ e $x2$ segundo a função linear: $$y = 2x_1-x_2-1 + \epsilon$$, onde $\epsilon$ é o ruído. A função linear que prevê valores de $y$ é dada por $$\hat{y} = 2x_1-x_2-1$$

O resultado é um gráfico de três dimensões:

````python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import random


fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111, projection='3d')

x1 = np.linspace(0, 10, 100)
x2 = np.linspace(-10, 5, 100)
y = [2 * i - j - 1 + random.normalvariate(mu=0, sigma=1) 
     for i, j in zip(x1, x2)]
y_ = [2 * i - j - 1 for i, j in zip(x1, x2)]

ax.scatter(x1, x2, y, s=100)
ax.plot(x1, x2, y_, c='r')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')
````

Para funções com mais de duas variáveis independentes não seria mais possível avaliar de maneira visual a qualidade da regressão, restando somente o coeficiente de determinação $R^2$ (```score()```) como medida de desempenho.

No exemplo das casas ```housing_ok.csv```, podemos criar um modelo de regressão múltipla para prever o preço dos imóveis com base na área contruída e a distância do aeroporto. Utilizaremos somente os imóveis cujo preço esteja entre 100 mil e 5 milhões de dólares e cuja área construída esteja entre 50 e 500 metros quadrados.

````python
import pandas as pd

df = pd.read_csv('dados/housing_ok.csv')
df = df[df.preco.between(1e5, 5e6)]
df = df[df.area_construida.between(50, 500)]
````

````python
from sklearn.linear_model import LinearRegression

X = df[['area_construida', 'distancia_aeroporto']].values
y = df.preco.values
reg = LinearRegression().fit(X, y)
reg.score(X, y)
````

Para o modelo, o valor de $R^2$ foi igual a 0,509 (50,9%). Esse valor pode ser interpretado como o percentual de preços que podem ser previstos a partir da área construída e a distância do aeroporto.

Abaixo, temos a visualização do modelo de regressão.

````python
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111, projection='3d')



ax.scatter(df.area_construida.values, 
           df.distancia_aeroporto.values, 
           df.preco.values, s=20, alpha=0.5)

x1 = np.linspace(df.area_construida.min(), df.area_construida.max(), 10000)
x2 = np.linspace(df.distancia_aeroporto.min(), df.distancia_aeroporto.max(), 10000)
X = np.append(x1.reshape(-1, 1), x2.reshape(-1, 1), axis=1)

ax.plot(x1, x2, reg.predict(X), c='red')
ax.set_xlabel('Área construída')
ax.set_ylabel('Distância até o aeroporto')
ax.set_zlabel('Preço do imóvel')
plt.title('Regressão linear múltipla')
````

Faremos, agora, um teste para a previsão do preço do imóvel com base nos atributos:
* ```area_construida```
* ```latitude```
* ```longitude```
* ```ano_construcao```

````python
X = df[['area_construida', 'latitude', 'longitude', 'ano_construcao']].values
y = df.preco.values
reg = LinearRegression().fit(X, y)
reg.score(X, y)
````

O resultado é um modelo com $R^2$ superior ao anterior. Faremos, agora, uma precisão a partir do último modelo criado.

````python
area_construida = 200
latitude = -37.74
longitude = 144.99
ano_construcao = 2015

reg.predict([[area_construida, latitude, longitude, ano_construcao]])
````
