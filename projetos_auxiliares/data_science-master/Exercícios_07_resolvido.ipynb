# Exercícios: Parte 07
O arquivo ```bmw.csv``` contem informações associadas a venda de de veículos usados da marca BMW no Reino Unido. Utilizando comandos de filtro de texto e agrupamento, gere as informações listadas a seguir:
1. Qual o preço (_price_) médio dos modelos (_model_) _Series_ (i.e. 1 _Series_, 2 _Series_, ...)? (utilize o comando ```str.contains()```)
2. Represente a distribuição de frequência dos preços somentes para os modelos Series.
3. Gere as estatísticas descritivas dos preços, agrupando por tipo de transmissão (_transmission_).
4. Gere as estatísticas descritivas dos preços, agrupando por modelo.
5. Gere a mesma informação anterior somente para os modelos Series, agrupando também pelo tipo de transmissão.
6. Descubra qual o modelo e ano que possui mais veículos a venda, indicando o preço médio para essa categoria.

````python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/ronaldolagepessoa/data_science/master/dados/bmw.csv')
df.head()
````

### Solução - 1

````python
df[df.model.str.contains('Series')].price.mean()
````

### Solução - 2

````python
df[df.model.str.contains('Series')].price.hist()
````

### Solução - 3

````python
df.groupby('transmission').price.describe()
````

### Solução - 4

````python
df.groupby('model').price.describe().sort_values('mean')
````

### Solução - 5

````python
df[df.model.str.contains('Series')].groupby(['model', 'transmission']).price.describe().sort_values('mean')
````

Solução - 6

````python
df[df.model.str.contains('Series')].groupby(['model', 'year']).price.describe().sort_values('count' ,ascending=False).head(1)
````
